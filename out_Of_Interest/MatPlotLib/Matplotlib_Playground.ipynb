```python
import matplotlib.pyplot as plt
```

```python
plt.plot([2,4,1,2,4,2,4,3,5,5,0],[5,1,8,8,1,7,8,2,4,4,5,])
plt.show()
```

```python
x1=([1,2,3,4])
y1=([2,5,3,4])

x2=([2,5,3,4])
y2=([2,5,2,3])

plt.plot(x1,y1, linestyle='-.',c='r',label='super line')
plt.plot(x2,y2, linestyle='-' ,c='g', label='hyper line')

plt.legend()

plt.title('Matplotlib Demo')

plt.xlabel('x axis')
plt.ylabel('y axis')

plt.show()
```
